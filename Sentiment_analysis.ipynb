```python
import pandas as pd
```

```python
ds_id = pd.read_csv('/content/drive/MyDrive/scaledata/Dennis+Schwartz/ds_id.csv')
ds_label3 = pd.read_csv('/content/drive/MyDrive/scaledata/Dennis+Schwartz/ds_label3.csv')
ds_label4 = pd.read_csv('/content/drive/MyDrive/scaledata/Dennis+Schwartz/ds_label4.csv')
ds_rating = pd.read_csv('/content/drive/MyDrive/scaledata/Dennis+Schwartz/ds_rating.csv')
ds_subj = pd.read_csv('/content/drive/MyDrive/scaledata/Dennis+Schwartz/ds_subj.csv')

jb_id = pd.read_csv('/content/drive/MyDrive/scaledata/James+Berardinelli/jb_id.csv')
jb_label3 = pd.read_csv('/content/drive/MyDrive/scaledata/James+Berardinelli/jb_label3.csv')
jb_label4 = pd.read_csv('/content/drive/MyDrive/scaledata/James+Berardinelli/jb_label4.csv')
jb_rating = pd.read_csv('/content/drive/MyDrive/scaledata/James+Berardinelli/jb_rating.csv')
jb_subj = pd.read_csv('/content/drive/MyDrive/scaledata/James+Berardinelli/jb_subj.csv')

scr_id = pd.read_csv('/content/drive/MyDrive/scaledata/Scott+Renshaw/scr_id.csv')
scr_label3 = pd.read_csv('/content/drive/MyDrive/scaledata/Scott+Renshaw/scr_label3.csv')
scr_label4 = pd.read_csv('/content/drive/MyDrive/scaledata/Scott+Renshaw/scr_label4.csv')
scr_rating = pd.read_csv('/content/drive/MyDrive/scaledata/Scott+Renshaw/scr_rating.csv')
scr_subj = pd.read_csv('/content/drive/MyDrive/scaledata/Scott+Renshaw/scr_subj.csv')

str_id = pd.read_csv('/content/drive/MyDrive/scaledata/Steve+Rhodes/str_id.csv')
str_label3 = pd.read_csv('/content/drive/MyDrive/scaledata/Steve+Rhodes/str_label3.csv')
str_label4 = pd.read_csv('/content/drive/MyDrive/scaledata/Steve+Rhodes/str_label4.csv')
str_rating = pd.read_csv('/content/drive/MyDrive/scaledata/Steve+Rhodes/str_rating.csv')
str_subj = pd.read_csv('/content/drive/MyDrive/scaledata/Steve+Rhodes/str_subj.csv')
```

```python
ds = [ds_id, ds_label3, ds_label4, ds_rating, ds_subj]
jb = [jb_id, jb_label3, jb_label4, jb_rating, jb_subj]
scr = [scr_id, scr_label3, scr_label4, scr_rating, scr_subj]
str0 = [str_id, str_label3, str_label4, str_rating, str_subj]
```

```python
ds1 = pd.concat(ds, ignore_index= True, axis =1, join = "outer")
jb1 = pd.concat(jb, ignore_index= True, axis =1, join = "outer")
scr1 = pd.concat(scr, ignore_index= True, axis =1, join = "outer")
str1 = pd.concat(str0, ignore_index= True, axis =1, join = "outer")
ds1
```

```python
ds1.columns = ['Id','Label3','Label4','Rating','Subject']
jb1.columns = ['Id','Label3','Label4','Rating','Subject']
scr1.columns = ['Id','Label3','Label4','Rating','Subject']
str1.columns = ['Id','Label3','Label4','Rating','Subject']
jb1
```

```python
print("DS1: ", ds1.shape)
print("JB1: ", jb1.shape)
print("SCR1: ", scr1.shape)
print("STR1: ", str1.shape)
```

```python
copy = ds1.append(jb1, ignore_index = True)
copy2 = scr1.append(str1, ignore_index = True)
df = copy.append(copy2, ignore_index = True)
df
```
